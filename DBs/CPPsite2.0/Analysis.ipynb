```python
import pickle
import pandas as pd
import seaborn
```

```python
with open('CPPSite2.pkl', 'rb') as fff:
    data_db = pickle.load(fff)
```

```python
data_db['1001']
```

```python
keys = set()
for ii in data_db:
    keys = keys | set(data_db[ii].keys())
```

```python
aux_dict = {ii:[] for ii in keys}
for ii,jj in data_db.items():
    for kk in keys:
        if kk in jj:
            aux_dict[kk].append(jj[kk])
        else:
            aux_dict[kk].append(None)

dataframe_db = pd.DataFrame(aux_dict)
```

```python
dataframe_db
```

```python
seaborn.histplot(dataframe_db['LENGTH'].astype(int), binwidth=1)
```
